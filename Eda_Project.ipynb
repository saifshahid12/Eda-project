**handling data :**

```python
# create a graph of colum floor and price
_ = df.groupby("floors")["price"].sum().plot(kind="bar")
plt.ylabel('total')
plt.title("determine floor from price")
plt.show()
```

**Plot a scatter plot :**

```python
# create scatter plot of two colums

sns.scatterplot(x="sqft_living",y="bedrooms",data=df,color='red')
plt.ylabel('y-axis')
plt.xlabel("x-axis")
plt.title("Living area")
plt.show()
```

```python
sns.histplot(df[["price", "bedrooms"]], color="red")
plt.ylabel('y-axis')
plt.xlabel("x-axis")
plt.title("Living area")
plt.show()
```

```python
# Prepare the values
values = [df["price"].sum(), df["bedrooms"].sum()]
labels = ["price", "bedrooms"]

# Plot
plt.pie(values, labels=labels, colors=["green"],autopct='%1.1f%%')

plt.ylabel('Bedrooms')
plt.xlabel("Price")
plt.title("Living area")
plt.show()
```

```python
values = [df["price"], df["bedrooms"], df['bathrooms'], df['floors'], df['waterfront'], df['sqft_living']]
labels = ["price", "bedrooms", "bathrooms", "floors", "waterfront", "sqft_living"]

# Plot
plt.boxplot(values, labels=labels)

plt.ylabel('y-axise')
plt.xlabel("x-axise")
plt.title("house graph")
plt.show()
```
