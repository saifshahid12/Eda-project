**Explanation 18 :**                               
                  
                  now i use pie chart graph of colums and sum after sum of
                  values it represent in the graph of their sum values , this
                  piechart is use for circle shape

**use piechart :**

```python
# Prepare the values
# use sum of colums
# make pie chart
# autopct is use for show percentage on graph

values = [df["price"].sum(), df["bedrooms"].sum()]
labels = ["price", "bedrooms"]

# Plot
plt.pie(values, labels=labels, colors=["green"],autopct='%1.1f%%')

plt.ylabel('Bedrooms')
plt.xlabel("Price")
plt.title("Living area")
plt.show()
```

**Explanation 19 :**                  
                   
                   now i draw a graph in boxplot and here i mention the label
                   because boxplot just give us plot not labet it , i manually
                   label this plot  and i compare the two colums in this graph
                   which is price and bedrooms here we add colors , price is
                   higher then bedrooms ,

**now use boxplot :**

```python
# Select the two columns of interest for their distributions
# plot the boxplot of tow colums

data_to_plot = [df['price'], df['bedrooms']]

# Create the box plot
plt.figure(figsize=(7, 6))
plt.boxplot(data_to_plot, labels=['Price', 'Bedrooms'], patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='blue'),
            medianprops=dict(color='red'),
            whiskerprops=dict(color='blue'),
            capprops=dict(color='blue'))

plt.title('Distribution of Price and Bedrooms')
plt.ylabel('Value')
plt.grid(axis='y', linestyle='--')
plt.show()
```
